```python
def check_for_category(diry, genre, img_data):
    filenames = np.array(os.listdir(diry))
    genre_filenames = set(img_data.loc[img_data['genre']==genre, ['filename']].values.flat)
    keepers = filenames[[fn in genre_filenames for fn in filenames]]
    
    counter = 0
    for k in keepers:
        subprocess.run('cp {}/{} data/{}/{}'.format(diry,k,genre,k).split())
        if counter % 100 == 0:
            print('{} images added.'.format(counter))
        counter += 1
        
    print('Done with {} images added'.format(counter))
    print('{} total images in {}'.format(len(os.listdir('data/{}'.format(genre))),genre))
    
```

```python
def create_dataset(path):
    filenames = np.array(os.listdir(path))
    imgs = []
    print('Creating dataset')
    for i in filenames:
        img = cv2.imread('{}/{}'.format(path,i))
        if img is not None:
            imgs.append(cv2.resize(img, (128,128), interpolation=cv2.INTER_AREA))
        if len(imgs)%100==0:
            print('{} images done.'.format(len(imgs)))
    
    print('{} images total.'.format(len(imgs)))
    print('Returning dataset')
    return np.array(imgs)
```
